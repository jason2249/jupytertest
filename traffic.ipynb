```python
with open("commute_times.txt") as f:
    lines = f.readlines()
print(lines)
```

```python
lines = lines[1:]
```

```python
print(lines)
```

```python
sum_numbers = [0 for i in range(24)]
for line in lines:
    split_line = line.split(",")
    time = int(split_line[2])
    hour = int(split_line[1])
    sum_numbers[hour] += time
print(sum_numbers)
```

```python
averages =[sum_num/365 for sum_num in sum_numbers]
print(averages)
```

```python
import matplotlib.pyplot as plt
plt.plot(averages)
plt.show()
```

```python
plt.plot([i for i in range(8,21)], averages[8:21])
plt.show()
```

```python
plt.xlabel("Hour of day")
plt.ylabel("Commute time in seconds")
plt.plot([i for i in range(8,21)], averages[8:21])
plt.show()
```
